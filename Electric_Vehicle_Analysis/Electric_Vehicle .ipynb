### Importing the neccessary Libary

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

### Raeding the file using pandas

```python
data = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

```python
# Display the first few rows of the dataframe
print(data.head())

# Display summary statistics of the dataframe
print(data.describe())

# Check for missing values in the dataframe
print(data.isnull().sum())

# Display the unique values in the 'Make' column
print(data['Make'].unique())

# Count the number of vehicles by 'Make'
make_counts = data['Make'].value_counts()
print(make_counts)


```

```python
# Plot the count of vehicles by 'Make'
plt.figure(figsize=(12, 8))
make_counts.plot(kind='bar')
plt.title('Count of Electric Vehicles by Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
