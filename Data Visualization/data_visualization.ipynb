## **PROJECT TWO**
### **Title:** Countries Data

### PROJECT _OBJECTIVES_

- To read the country data file using pandas
- To compare the two Country population (_us and China_ ) and visualized the results.
- To compare the population growth by 100

### _Importing the country Data using pandas_

```python
countries_data = pd.read_csv('countries.csv')
countries_data
```

### _COMPARING THE POPULATION IN US AND CHINA_

```python
us = countries_data[countries_data.country == 'United States']
us
#Explanation 
# When you write countries_data == 'united States'--- This will return only when the country is True
#When you attached countries_data---meaning a portion of the country data
```

```python
china = countries_data[countries_data.country == 'China']
china
```

### _Visualizing the results_

```python
#Plot Population of the two country
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)      #Adding China to the plot
plt.legend(['United States', 'China'])              #Adding legend
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()                                       #Showing the plot
```

### _Showing Population Growth of the two countries by 100_

```python
us.population / us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)      
plt.legend(['United States', 'China'])              
plt.xlabel("Year")
plt.ylabel("Population Growth {first year = 100}")
plt.show()         
```
